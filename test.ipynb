```python
import pandas as pd #加入pandas套件進行資料處理
import numpy as np #加入numpy套件進行數值運算

df = pd.read_csv("Pokemon.csv") #讀取Pokemon資料集(確認檔案路徑正確)


misscounts = df.isnull().sum() #isnull先檢查每個欄位的缺失值殘缺回傳true(1)，sum計算缺失值總數就是計算那個1
print("各欄位缺失值數量：")
print(misscounts[misscounts > 0]) #輸出缺失值數量大於0的欄位

# 顯示 Type 2 曾為缺失值 (現在是 'None') 的前幾筆資料
print(df[df['Type 2'] == 'None'].head().to_markdown(index=False, numalign="left", stralign="left"))


# 策略：使用字串 'None' 來填補 Type 2 欄位中的所有缺失值 (NaN)
df['Type 2'] = df['Type 2'].fillna('None')

# 再次檢查 Type 2 欄位是否還有缺失值
print("Type 2 欄位處理後的缺失值數量：")
print(df['Type 2'].isnull().sum())
```
